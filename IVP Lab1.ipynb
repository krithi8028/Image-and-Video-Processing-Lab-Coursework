```python
import numpy as np

import matplotlib.pyplot as plt

from PIL import Image, ImageDraw, ImageFont
```

```python
# Load and display an image

image_path = "D:/Backup/Downloads/Vadivelu.jpg"  # Example image URL

image = Image.open(image_path)

plt.imshow(image)

plt.title("Original Image")

plt.axis('off')

plt.show()
```

```python
# Convert the image to grayscale

gray_image = image.convert("L")

plt.imshow(gray_image, cmap='gray')

plt.title("Grayscale Image")

plt.axis('off')

plt.show()
```

```python
# Resize the image

resized_image = image.resize((150, 150))

plt.imshow(resized_image)

plt.title("Resized Image")

plt.axis('off')

plt.show()
```

```python
# Crop a region of interest (ROI) from the image

roi = image.crop((300, 100, 500, 200))  # (left, upper, right, lower)

plt.imshow(roi)

plt.title("Cropped ROI")

plt.axis('off')

plt.show()
```

```python
# Save the modified image
roi.save("cropped_image.png")
```

```python
# Rotate the image by 90, 180, and 270 degrees

rotated_90 = image.rotate(90)

rotated_180 = image.rotate(180)

rotated_270 = image.rotate(270)


plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)

plt.imshow(rotated_90)

plt.title("Rotated 90 Degrees")

plt.axis('off')


plt.subplot(1, 3, 2)

plt.imshow(rotated_180)

plt.title("Rotated 180 Degrees")

plt.axis('off')


plt.subplot(1, 3, 3)

plt.imshow(rotated_270)

plt.title("Rotated 270 Degrees")

plt.axis('off')


plt.show()
```

```python
# Flip the image horizontally and vertically

flipped_horizontal = image.transpose(method=Image.FLIP_LEFT_RIGHT)

flipped_vertical = image.transpose(method=Image.FLIP_TOP_BOTTOM)


plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)

plt.imshow(flipped_horizontal)

plt.title("Flipped Horizontally")

plt.axis('off')


plt.subplot(1, 2, 2)

plt.imshow(flipped_vertical)

plt.title("Flipped Vertically")

plt.axis('off')


plt.show()
```

```python
# Draw basic shapes and add text on the image

draw = ImageDraw.Draw(image)

draw.line((0, 0) + image.size, fill=128, width=5)

draw.line((0, image.size[1], image.size[0], 0), fill=128, width=5)

draw.rectangle([150, 150, 100, 100], outline="red", width=3)

draw.ellipse([200, 200, 250, 250], outline="blue", width=3)

draw.text((50, 50), "Vannakamga", fill="white")


plt.imshow(image)

plt.title("Image with Shapes and Text")

plt.axis('off')

plt.show()
```

```python

# Change the image color intensities by scaling the RGB channels independently

r, g, b = image.split()

r = r.point(lambda i: i * 1.5)  # Scale red channel

g = g.point(lambda i: i * 0.5)   # Scale green channel

b = b.point(lambda i: i * 1.0)   # Scale blue channel

scaled_image = Image.merge("RGB", (r, g, b))


plt.imshow(scaled_image)

plt.title("Color Intensities Adjusted")

plt.axis('off')

plt.show()
```

```python
# Create a collage by combining multiple images of the same size into a single grid

collage_width = 300

collage_height = 300

collage = Image.new('RGB', (collage_width * 2, collage_height * 2))


collage.paste(image.resize((collage_width, collage_height)), (0, 0))

collage.paste(resized_image, (collage_width, 0))

collage.paste(roi.resize((collage_width, collage_height)), (0, collage_height))

collage.paste(scaled_image.resize((collage_width, collage_height)), (collage_width, collage_height))


plt.imshow(collage)

plt.title("Image Collage")

plt.axis('off')

plt.show()
```
